```python
import pandas as pd
from lets_plot import *
LetsPlot.setup_html()
```

```python
ldepth=pd.DataFrame()
idepth=pd.DataFrame()
imiss=pd.DataFrame()
lmiss=pd.DataFrame()
frq=pd.DataFrame()
lqual=pd.DataFrame()
```

```python
for option in ["", "options"]:
    for qc in ["NoQC", "QC"]:
        ldepth_tmp=pd.read_table("/home/share/GBS_honeybee/Run2/Stacks_pipeline/5_population/" + option + "/" + qc + "/stats/populations.snps.vcf.ldepth.mean")
        ldepth_tmp.loc[:, "option"] = option
        ldepth_tmp.loc[:, "qc"] = qc
        idepth_tmp=pd.read_table("/home/share/GBS_honeybee/Run2/Stacks_pipeline/5_population/" + option + "/" + qc + "/stats/populations.snps.vcf.idepth")
        idepth_tmp.loc[:, "option"] = option
        idepth_tmp.loc[:, "qc"] = qc
        lmiss_tmp=pd.read_table("/home/share/GBS_honeybee/Run2/Stacks_pipeline/5_population/" + option + "/" + qc + "/stats/populations.snps.vcf.lmiss")
        lmiss_tmp.loc[:, "option"] = option
        lmiss_tmp.loc[:, "qc"] = qc
        imiss_tmp=pd.read_table("/home/share/GBS_honeybee/Run2/Stacks_pipeline/5_population/" + option + "/" + qc + "/stats/populations.snps.vcf.imiss")
        imiss_tmp.loc[:, "option"] = option
        imiss_tmp.loc[:, "qc"] = qc
        frq_tmp=pd.read_table("/home/share/GBS_honeybee/Run2/Stacks_pipeline/5_population/" + option + "/" + qc + "/stats/populations.snps.vcf.frq")
        frq_tmp.loc[:, "option"] = option
        frq_tmp.loc[:, "qc"] = qc
        lqual_tmp=pd.read_table("/home/share/GBS_honeybee/Run2/Stacks_pipeline/5_population/" + option + "/" + qc + "/stats/populations.snps.vcf.lqual")
        lqual_tmp.loc[:, "option"] = option
        lqual_tmp.loc[:, "qc"] = qc
        ldepth = pd.concat([ldepth, ldepth_tmp])
        idepth = pd.concat([idepth, idepth_tmp])
        lmiss = pd.concat([lmiss, lmiss_tmp])
        imiss = pd.concat([imiss, imiss_tmp])
        frq = pd.concat([frq, frq_tmp])
        lqual = pd.concat([lqual, lqual_tmp])
        
```
